```python
import pandas as pd
from sklearn import model_selection, preprocessing, linear_model, naive_bayes, metrics, svm
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics import f1_score
```

### CountVectorizer tweet_lemmatized

```python
count_lem=pd.read_csv('count_lem.csv', index_col=0)
```

```python
y=count_lem['label']
```

```python
count_lem.drop(labels='label',axis=1, inplace=True )
```

```python
x_train,x_valid,y_train,y_valid=model_selection.train_test_split(count_lem[:31962],y[:31962])
```

```python
lr=linear_model.LogisticRegression()
lr.fit(x_train,y_train)
```

```python
answer=lr.predict_proba(x_valid)
```

```python
y_pred=[1 if i[1]>=0.3 else 0 for i in answer ]
```

```python
metrics.f1_score(y_valid,y_pred)
```
